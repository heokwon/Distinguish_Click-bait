# 원문, 추출요약, 생성요약 가져오기

```python
import pandas as pd
import json
```

```python
with open('/content/drive/MyDrive/문서요약/json/train_original3.json', 'r') as j:
  contents = json.loads(j.read())
```

```python
train_sent = []

for document in contents['documents']:
  for text in document['text']:
    sents = ''
    for tex in text:
      sent = tex['sentence'] 
      
      sents = sents + sent
      print(sents)
    train_sent.append(sents)
  break
```

## 제목 뽑기

```python
train_titles = []

for document in contents['documents']:
  title = document['title']
  train_titles.append(title)
```

## 원문 뽑기

```python
train_sents = []

for document in contents['documents']:
  sents = ''
  for text in document['text'][2:]:
    sent_sub = ''
    for tex in text:
      
      
      sent = tex['sentence']
      
      sent_sub += sent
    sents += sent_sub
  train_sents.append(sents)
  
```

```python
len(train_sents)
```

```python
train_sents[:5]
```

## 생성요약 뽑기

```python
train_ab_sents = []

for document in contents['documents']:
  ab_sent = document['abstractive'][0]
  train_ab_sents.append(ab_sent)
```

```python
print(len(train_ab_sents)) 
train_ab_sents[:5]
```

## 추출요약 뽑기

```python
ex_index_lists = []

for document in contents['documents']:
  ex_index_lists.append(document['extractive'])
```

```python
len(ex_index_lists)
```

```python
ex_index_lists[:5]
```

```python
train_ex_sents = []

for document, idx in zip(contents['documents'], ex_index_lists):
  
  
  sent_sub = ''
  for text in document['text'][2:]:
    
    for tex in text:
      
      if tex['index'] in idx:
         sent = tex['sentence']
         sent_sub += sent

      
      
  train_ex_sents.append(sent_sub)
  
     
      
      
      
```

```python
len(train_ex_sents)
```

```python
train_ex_sents[:5]
```

```python
df = pd.DataFrame(zip(train_titles, train_sents, train_ex_sents, train_ab_sents), columns = ['제목', '원문', '추출요약', '생성요약'])
```

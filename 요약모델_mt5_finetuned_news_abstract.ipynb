```python
from transformers import AutoTokenizer, TFAutoModelForSeq2SeqLM


tokenizer = AutoTokenizer.from_pretrained("jiseong/mt5-small-finetuned-news-ab")
model = TFAutoModelForSeq2SeqLM.from_pretrained("jiseong/mt5-small-finetuned-news-ab")
```

```python
from transformers import SummarizationPipeline

summarizer = SummarizationPipeline(model=model, tokenizer=tokenizer, framework="tf")
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/문서요약/신문기사.csv')
df
```

```python
for i in range(10):

    text = df['원문'][(i*10)]
    result = summarizer(text, min_length=5, max_length=300)
    label = df['생성요약'][(i*10)]   

    print('==='*50)
    print(f'원문 : {text}')
    print(f"요약문 : {result[0]['summary_text']}")
    print(f"정답 : {label}")
    print('==='*50)
```

```python
text_list = []
result_list = []
label_list = []

for i in range(10):

    text = df['원문'][100000+(i*10)]
    result = summarizer(text, min_length=5, max_length=300)
    label = df['생성요약'][100000+(i*10)]   
    
    text_list.append(text_list)
    result_list.append(result[0]['summary_text'])
    label_list.append(label)

    print('==='*50)
    print(f'원문 : {text}')
    print(f"요약문 : {result[0]['summary_text']}")
    print(f"정답 : {label}")
    print('==='*50)
```

```python
from datasets import load_metric

metric = load_metric("rouge")
```

```python
for result, label in zip(result_list, label_list):
    print(metric.compute(predictions=[result], references=[label]))
    
```
